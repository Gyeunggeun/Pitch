```python
from bs4 import BeautifulSoup
import requests

# 페이지 URL
url = 'https://www.lgtwins.com/service/html.ncd?baRq=IN_DS&baRs=OUT_DS%2COUT_CAREER_DS&IN_DS.USE_YN=Y&TYPE=PLYPTC&view=%2Fpc_twins%2Ftwins_player%2Ftwins_player_list&actID=BR_RetrieveTeamMemberInfo'

# 페이지의 HTML 가져오기
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# 사진이 있는 모든 태그 찾기
image_tags = soup.find_all('img')

# 각 태그에서 src 속성(이미지 URL) 가져오기
image_urls = [img['src'] for img in image_tags if 'src' in img.attrs]

# 결과 출력
for url in image_urls:
    print(url)
```

```python
import os
from urllib.parse import urljoin
import requests

# 기본 URL 설정
base_url = 'https://www.lgtwins.com'

# 이미지 저장 경로 설정
path = './crawled_images'

# 경로가 없으면 생성
if not os.path.isdir(path):
    os.mkdir(path)

# 각 이미지 URL에 대하여
for idx, url in enumerate(image_urls):
    # 상대 URL을 절대 URL로 변환
    url = urljoin(base_url, url)

    # 이미지 데이터 가져오기
    response = requests.get(url)

    # 파일 경로 생성
    filename = os.path.join(path, f'image_{idx}.jpg')

    # 이미지 저장
    with open(filename, 'wb') as f:
        f.write(response.content)
```
