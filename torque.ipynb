```python
import pandas as pd
```

```python
df = pd.read_csv("torque/20230612020119_2.csv")
df.head()
```

```python
# sqrt((phi^^)^2 + (theta^^)^2 + (psi^^)^2) 으로 새로운 열 추가
df['omega'] = (df['phi^']**2 + df['theta^']**2 + df['psi^']**2)**0.5
df.head()
```

```python
# sqrt((phi^^)^2 + (theta^^)^2 + (psi^^)^2) 으로 새로운 열 추가
df['omega2'] = (df['phi^^']**2 + df['theta^^']**2 + df['psi^^']**2)**0.5
df.head()
```

사용 공식

\begin{align*}
a_{elbow} &= A - \dot{\omega} \times L - \omega \times (\omega \times L) - g \\
a_{com} &= a_{elbow} + COM(\dot{\omega} \times L) + \omega \times COM(\omega \times L) \\
T_{elbow} &= I_{forearm} \dot{\omega} + COM[L \times (m_{forearm} a_{com} - m_{forearm} g)]
\end{align*}


가정값

\begin{align*}
A = 287.07 m/s² \\

I_{forearm} = 0.02 kg.m²\\

COM = 0.3384m \\

g = 9.8m/s² \\

L = 0.29m \\

m_{forearm} = 3.5kg \\ 
\end{align*}

```python
# a elbow 열 추가
df['a elbow'] = 287.07 - df['omega'] * 0.5 - df['omega2']**2 * 0.29 - 9.8
```

```python
df['a com'] = df['a elbow'] + 0.3384 * df['omega2']**2 * 0.29 + 0.29 * df['omega']**2 * 0.3384
```

```python
df['torque'] = 0.02 * df['omega2'] + 0.3384 * 0.29 * (df['a com'] - 9.8) * 2
```
